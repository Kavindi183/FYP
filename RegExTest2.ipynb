```python
# a number following 'dengue' or 'cases' or 'suspected dengue cases' or 'suspected cases'
regex_for_cases = re.compile(r'(?:\d+(?:(?:,|\s|(?:,\s)|(?:\s,))\d+)*\s*(?:dengue|cases|suspected\sdengue\scases|suspected\+*-scases))')
```

```python
regex_for_deaths1 = re.compile(r'\d+(?:(?:,|\s|(?:,\s)|(?:\s,))\d+)*\s*(?:deaths|death|people\shave\sdied)')
```

```python
regex_for_deaths = re.compile(r'\d+(?:(?:,|\s|(?:,\s)|(?:\s,))\d+)*\s*(?:deaths|death|(?:people\s|patients\s)?(?:have|had)\sdied)')
```

```python
regex_for_dates = re.compile(r'(?:\d{4}(?:\.|/|-)\d{1,2}(?:\.|/|-)\d{1,2})|(?:\d{1,2}(?:\.|/|-)\d{1,2}(?:\.|/|-)\d{2,4})|(?:(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|April|May|June|July|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\s\d{1,2})')
```

```python
def get_deaths(text):
    return regex_for_deaths.findall(text)       
```

```python
def get_numbers(text):
    return regex_for_numbers.findall(text)
```

```python
def get_cases(text):
    return regex_for_cases.findall(text)
```

```python
def get_published_date(text):
    return re.search(regex, text).group()
```

```python
def get_dates(text):
    return regex_for_dates.findall(text)
```
